```python
import pandas as pd
import numpy as np
import math
from scipy import stats
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
from matplotlib import pyplot as plt
```

```python
data_new = pd.melt(data.reset_index(),id_vars=['index'],value_vars=["Teaching_Method1","Teaching_Method_2","Teaching_Method_3"])
data_new.columns=['index','Treatment','value']
```

```python
data_new
```

```python
model=ols('value ~ C(Treatment)', data = data_new).fit()
```

```python
anova_table=sm.stats.anova_lm(model,typ=1)
```

```python
anova_table
```

```python
# ANOVA - CONCEPTUAL OVERVIEW
# H0 : u1 = u2 = u3 = u4 ==== uk
# Ha : u1 != u2 != u3 != u4 !=!=!=!= uk 

# Assumption of ANOVA
#  fOR Each population , the dependant variable is normally distributed
# The variance of the dependant variable (sima^2), is the same for  all the populations

# the observations must be independent
```
