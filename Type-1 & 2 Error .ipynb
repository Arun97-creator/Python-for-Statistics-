```python
from scipy import stats
def z_value(x,mu,SEM):
    z = (x- mu)/SEM
    if(z<0):
        alfa = stats.norm.cdf(z)
    else:
        alfa = 1- stats.norm.cdf(z)
    print(alfa)    
```

```python
#Calculating alpha for different values of x, mu and SEM
x = 48.5
mu = 50
SEM = 0.79
```

```python
z_value(x,mu,SEM)
```

```python
import numpy as np
def type_2(mu1,mu2,sigma,n,alfa):
    z =stats.norm.ppf(alfa)
    xbar = (mu1)+(z*sigma/np.sqrt(n))
    z2 = (xbar - mu2)/(sigma/np.sqrt(n))
    if(mu1 > mu2):
        beta= 1-stats.norm.cdf(z2)
    else:
        beta = stats.norm.cdf(z2)
    print(beta)
    print(xbar)
       
# Here first we find xbar value , z@ 5% significance= 1.64 ( through which we will define our acceptance region)
#Eerything on its left is rejection , in type 2 we find a new mean =7.4 and find the prob of right area which has the chances 
# of not being rejected
```
